```python
# Data Visualization
# Project 4
# Author: Kevin Zeng

import pandas as pd
import plotly.express as px

cd_df = pd.read_csv('crude.csv')

fig = px.line(cd_df, x="date", y="price", title='Crude Oil Price from 1995 to Present', markers=True)
fig.update_layout(
    xaxis_title="Date",
    yaxis_title="Crude Oil Price (USD/BBL)"
)

fig.update_traces(marker=dict(
    color='green'))

fig.show()
```
